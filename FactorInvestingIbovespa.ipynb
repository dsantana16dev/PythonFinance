```python
# ver as cotações

dados_cotacoes = (yf.download(lista_acoes['tickers'].to_list(), start="2015-06-30", end="2022-12-31")['Adj Close'])

dados_cotacoes
```

### Tratar os dados

```python
# transformar o indice em data
dados_cotacoes.index = pd.to_datetime(dados_cotacoes.index)

dados_cotacoes
```

```python
# sort_index() para ordenar a planilha
dados_cotacoes = dados_cotacoes.sort_index()
dados_cotacoes
```

### calcular o retorno 6 meses

```python
# resample() <- remonta a serie temporal / pct_cahnge <- usado para obter a alteração percentual entre os dados.
# axis->0 Linha axis->1 coluna / how="all" toda a linha com NaN
r6 = (dados_cotacoes.resample("M").last().pct_change(periods = 6).dropna(axis=0, how="all").drop("2022-12-31"))
r6
```

### Classificar e retirar empresas que não participaram IBOVE

```python
# localizar item .loc

r6.loc['2015-12-31', 'ABEV3.SA']
```

```python
#exemplo replace()
empresa_teste = "AMBEV.SA"
print(empresa_teste.replace(".SA", ""))
```

```python
for data in r6.index:
    for empresa in r6.columns:
        
        if empresa.replace(".SA", "") not in composicao_historica.loc[:, data].to_list():
        
            r6.loc[data, empresa] = pd.NA
            
r6
```

```python
# Exemplo simples de função Lambda -> pega todos os elemento e aplica algo

desconto = 0.3

preco_das_malas = [50, 100, 290]

desconto_das_malas = list(map(lambda x: x * desconto, preco_das_malas))

desconto_das_malas
```

### Carteira de Investimento

```python
# fazer ranking - ver quais foram as melhores empresas
rankings = r6.rank(axis = 1, ascending=False)
rankings
```

```python
# fazer a carteira com apemas as 15 melhores ações da IBOVE

carteiras = r6.rank(axis = 1, ascending = False).applymap(lambda x: 1 if x < 11 else 0)

carteiras
```

### Calcular retorno mensal

```python
retorno_mensal = dados_cotacoes.resample("M").last().pct_change(periods = 1)

retorno_mensal = retorno_mensal.drop(retorno_mensal.index[:7], axis = 0)

carteiras.index = retorno_mensal.index

retorno_mensal
```

## retorno mensal X carteira

```python
retorno_modelo = (retorno_mensal * carteiras).sum(axis = 1)/10

retorno_modelo
```

## Visualizar a rentabilidade

```python
qs.extend_pandas()

retorno_modelo.plot_monthly_heatmap()
```

## comparar modelo Ibove

```python
ibovespa = yf.download("^BVSP", start = "2015-12-30", end = "2022-12-31")['Adj Close']

ibovespa = ibovespa.resample("M").last().pct_change().dropna()

ibovespa
```

```python
#retorno acumulado

retorno_acum_modelo = (1 + retorno_modelo).cumprod() - 1 
retorno_acum_ibov =  (1 + ibovespa).cumprod() - 1

retorno_acum_modelo.plot_monthly_heatmap()
retorno_acum_ibov.plot_monthly_heatmap()
```

## Retorno do MODELO vs IBOVESPA

```python
ganhos_modelo = retorno_modelo - ibovespa

ganhos_modelo.plot_monthly_heatmap()
```
